# Fill blanks

```python
train_test=[train, test]
```

```python
for dataset in train_test: dataset["Title"]= dataset["Name"].str.extract("([\w]+)\.", expand=False)
```

```python
train['Title'].value_counts()
```

```python
test['Title'].value_counts()
```

```python
Title_mapping= {"Mr": 0, "Miss":1, "Mrs":2, "Master":3, "Dr":4, "Rev":5,"Mlle":6, "Major":7,
                "Col":8,"Countess":9, "Capt":10, "Ms":11, "Sir":11, "Lady":12, "Mme":13, "Don":14,
                "Jonkheer":15, "Dona":16, "Dr":17}
for dataset in train_test:
    dataset["Title"]=dataset["Title"].map(Title_mapping)
```

```python
train["Title"].unique()
```

```python
test["Title"].unique()
```

```python
for dataset in train_test: dataset["Title_group"]= dataset["Name"].str.extract("([\w]+)\.", expand=False)

for dataset in train_test:
    dataset['Title_group'] = dataset['Title_group'].apply(lambda x: 0 if x=="Mr" else 1 if x=="Miss" else 2 if x=="Mrs" else 3 if x=="Master" else 4)
```

```python
train.head()
```

```python
test.head()
```

```python
for dataset in train_test: dataset['Age'].fillna(dataset.groupby("Title_group")["Age"].transform("median"), inplace=True)
```

```python
for dataset in train_test: dataset["Fare"].fillna(dataset.groupby("Pclass")["Fare"].transform("median"), inplace=True)
```

# Replace categorical values to binary

```python
# Sex
sex_mapping = {"male": 0, "female":1}
for dataset in train_test:
    dataset["Sex"]=dataset["Sex"].map(sex_mapping)
```

```python
# Embarked
dataset["Embarked"].unique()
```

```python
Embarked_mapping = {"C":0, "Q":1, "S":2}
for dataset in train_test:
    dataset["Embarked"]=dataset["Embarked"].map(Embarked_mapping)
```

```python
for dataset in train_test: dataset['Embarked'].fillna(dataset.groupby("Title_group")["Embarked"].transform("median"), inplace=True)
```

# Specific grouping

```python
for dataset in train_test:
    dataset["Agebin"]= pd.cut(dataset["Age"], 5, labels= [0,1,2,3,4])
```

```python
for dataset in train_test:
    dataset["Farebin"]= pd.cut(dataset["Fare"], 4, labels= [0,1,2,3])
```

# Create the answer sheet and final train data

```python
drop_column=["Name", "Age", "SibSp","Parch","Ticket","Fare","Cabin", "Title"]

for dataset in train_test:
    dataset=dataset.drop(drop_column, axis=1, inplace=True)
```

```python
target= train["Survived"]
target
```

```python
drop_column2=["PassengerId", "Survived"]
train_data=train.drop(drop_column2, axis=1)
train_data
```

# Testing the Models

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
# Initialize Logistic Regression model
clf1 = LogisticRegression()

# Fit the model on the training data
clf1.fit(train_data, target)

# Check accuracy on the training data
accuracy = clf1.score(train_data, target)
print("Accuracy:", accuracy)
```

```python
## Initialize Random forest model
clf2 = RandomForestClassifier()

# Fit the model on the training data
clf2.fit(train_data, target)

# Check accuracy on the training data
accuracy = clf2.score(train_data, target)
print("Accuracy:", accuracy)
```

```python
# Initialize Gaussian model
clf3 = GaussianNB()

# Fit the model on the training data
clf3.fit(train_data, target)

# Check accuracy on the training data
accuracy = clf3.score(train_data, target)
print("Accuracy:", accuracy)
```

```python
# Initialize Decision Tree model
clf4 = DecisionTreeClassifier()

# Fit the model on the training data
clf4.fit(train_data, target)

# Check accuracy on the training data
accuracy = clf4.score(train_data, target)
print("Accuracy:", accuracy)
```

```python
# Initialize SVC model
clf5 = SVC()

# Fit the model on the training data
clf5.fit(train_data, target)

# Check accuracy on the training data
accuracy = clf5.score(train_data, target)
print("Accuracy:", accuracy)
```

# Create the final test data
